```python
import pandas as pd
import networkx as nx

# Baca file
file1 = pd.read_csv("mcr_similarity_network.csv")
file2 = pd.read_csv("expanded_network.csv")

# Gabung file
file_master = pd.concat([file1, file2])

# Buat jaringannya
G = nx.from_pandas_edgelist(file_master, source='source', target='target')

print(f"Jaringan kamu sekarang punya {G.number_of_nodes()} artis!")
```

```python
# Menghitung jumlah koneksi tiap artis
derajat_koneksi = dict(G.degree())

# Urutkan dari yang paling banyak koneksinya
ranking_artis = sorted(derajat_koneksi.items(), key=lambda x: x[1], reverse=True)

print("Artis paling populer di jaringan kamu:")
for artis, jumlah in ranking_artis[:5]:
    print(f"- {artis}: punya {jumlah} koneksi")
```

```python
from pyvis.network import Network

# 1. Inisialisasi Network Pyvis
# notebook=True khusus untuk jalan di dalam .ipynb
net = Network(height="95vh", width="100%", bgcolor="#222222", font_color="white", notebook=True)

# 2. Masukkan data dari NetworkX (G) ke Pyvis
net.from_nx(G)

# 3. Tambahkan fitur interaktif (Physics)
net.toggle_physics(True)

# 4. Simpan ke HTML
print("Sedang membuat file network_interactive.html...")
net.show("network_interactive.html")
```
