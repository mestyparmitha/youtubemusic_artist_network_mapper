```python
import csv
import time
from ytmusicapi import YTMusic

# Inisialisasi API (Tanpa login pun bisa untuk data publik)
yt = YTMusic()
```

```python
def crawl_artis_circle(seed_name="My Chemical Romance", limit=15):
    print(f"Mencari data untuk: {seed_name}")
    
    # 1. Cari artis utama
    search = yt.search(seed_name, filter="artists")
    if not search:
        print("Artis tidak ditemukan.")
        return
    
    artist_id = search[0]['browseId']
    artist_data = yt.get_artist(artist_id)
    main_artist_name = artist_data['name']
    
    # 2. Ambil related artists
    related = artist_data.get('related', {}).get('results', [])
    print(f"Berhasil menemukan {len(related)} artis terkait.\n")
    
    # 3. Simpan ke csv
    network_results = []
    
    for i, artist in enumerate(related[:limit]):
        target_name = artist['title']
        target_id = artist['browseId']
        
        # simpan relasi
        network_results.append({
            'source': main_artist_name,
            'target': target_name,
            'target_id': target_id,
            'relationship': 'Similar',
        })
        print(f"[{i+1}] Menghubungkan {main_artist_name} ➔ {target_name}")
        
    # 4. Simpan hasil ke CSV
    filename = "mcr_similarity_network.csv"
    keys = network_results[0].keys()
    
    with open(filename, 'w', newline='', encoding='utf-8') as f:
        dict_writer = csv.DictWriter(f, fieldnames=keys)
        dict_writer.writeheader()
        dict_writer.writerows(network_results)
        
    print(f"Selesai! Data disimpan di: {filename}")

if __name__ == "__main__":
    crawl_artis_circle()
```
